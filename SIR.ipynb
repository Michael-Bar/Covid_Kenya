```python
import sys, datetime, io
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

Fetch real world up-to-date data

```python
url = "https://www.ecdc.europa.eu/sites/default/files/documents/COVID-19-geographic-disbtribution-worldwide-{}.xlsx".format(datetime.datetime.now().date())
url

```

```python
df = pd.read_excel(url,
                 sep='\t',
                 parse_dates=[0],
)
```

```python
df['DateRep'] = pd.to_datetime(df.DateRep)
df.columns = df.columns.str.replace(" ","_")
df.head(3)
```

Fill in blank dates:

```python
dfs = []
for country in df.Countries_and_territories.unique():
    d = df[df.Countries_and_territories==country].copy()
    d.set_index('DateRep',inplace=True)
    idx = pd.date_range(d.index.min(), d.index.max())
    d = d.reindex(idx, fill_value=0)
    d['DateRep'] = d.index
    d.reset_index(inplace=True, drop=True)
    dfs.append(d)
    
df = pd.concat(dfs)
df.reset_index(inplace=True, drop=True)
```

```python
def normalise_country(dataf, Country='China', cases=15, case_type = 'reported'):
    if case_type is 'estimated':
        df_equiv = dataf.query("Countries_and_territories == '{}' and est_real_cases_cumsum> {}".format(Country, cases)).sort_values("DateRep")
        df_equiv['normalised_date'] = df_equiv.doy - np.min(df_equiv.doy)
    elif case_type is "reported":
        df_equiv = dataf.query("Countries_and_territories == '{}' and cum_cases> {}".format(Country, cases)).sort_values("DateRep")
        df_equiv['normalised_date'] = df_equiv.doy - np.min(df_equiv.doy)
    return df_equiv
    
```

```python
df['doy'] = df.DateRep.dt.dayofyear
df['doy'] = [df.loc[i].doy-365 if df.loc[i].Year==2019  else df.loc[i].doy for i in df.index ]
dfs = [df[df.Countries_and_territories==i] for i in df.Countries_and_territories.unique()]
```

```python
df = []
for country in dfs:
    country = country.sort_values("DateRep", ascending=True)
    country['cum_cases'] = country.Cases.cumsum()
    df.append(country)
df = pd.concat(df)
```

# Estimated real cases

assuming a 1% mortality rate and 14d incubation we can calculate what the estimated *actual* infection was (as the detected vs infected depends on the number of tests completed, and we therefore assume the number of deaths is a more accurate number)

```python
dfs = []
for country in df.Countries_and_territories.unique():
    d = df[df.Countries_and_territories==country]
    d['est_real_cases'] = d.Deaths.shift(-14)
    d['est_real_cases'] = d.est_real_cases * 100.
    d['est_real_cases_cumsum'] = d.est_real_cases.cumsum()
    dfs.append(d)
df = pd.concat(dfs)
```

## SIR model 

```python
S = 50e6 # population init

N_hospitalized = 3 #number in hospital now
hospitalization_rate = 5 # percent of people needing hosp, estimate from lit

initial_infections = 15 #known (detected) cases

relative_contact_rate = 0 # 0-100% effectiveness of lockdown

dt = 6.4 #doubling time in days, from lit: https://www.thelancet.com/journals/lancet/article/PIIS0140-6736(20)30260-9/fulltext

recovery_days = 14.0 #est from lit
```

```python


hospitalization_rate = hospitalization_rate/100. # convert
relative_contact_rate = relative_contact_rate/100. # convert

real_infections = initial_infections #N_hospitalized/hospitalization_rate # estimate real total from hospitalized 
detection_prob = initial_infections / real_infections #estimate detection probability


S, I, R = S, initial_infections / detection_prob, 0 #ie based on latest data inputted above


intrinsic_growth_rate = 2 ** (1 / dt) - 1

gamma = 1 / recovery_days

# Contact rate, beta
beta = (
    intrinsic_growth_rate + gamma
) / S * (1-relative_contact_rate) # {rate based on doubling time} / {initial S}


r_t = beta / gamma * S # r_t is r_0 after distancing
r_naught = r_t / (1-relative_contact_rate)
doubling_time_t = 1/np.log2(beta*S - gamma +1) # doubling time after distancing
```

```python
print("inputs: \n \t Detected cases: {} \n \t hospitlization rate {}, \n \t real cases: {} \n \t doubling time {} \n \t recovery_days".format(initial_infections, hospitalization_rate, real_infections,dt, recovery_days))


```

```python
print("Outputs: \n \t R0: {} ".format(r_naught))


```

```python
print("Lancet reported R0=2.7") #https://www.thelancet.com/journals/lancet/article/PIIS0140-6736(20)30260-9/fulltext
```

```python
def sir(y, beta, gamma, N):
    S, I, R = y
    Sn = (-beta * S * I) + S
    In = (beta * S * I - gamma * I) + I
    Rn = gamma * I + R
    if Sn < 0:
        Sn = 0
    if In < 0:
        In = 0
    if Rn < 0:
        Rn = 0

    scale = N / (Sn + In + Rn)
    return Sn * scale, In * scale, Rn * scale


# Run the SIR model forward in time
def sim_sir(S, I, R, beta, gamma, n_days, beta_decay=None):
    N = S + I + R
    s, i, r = [S], [I], [R]
    for day in range(n_days):
        y = S, I, R
        S, I, R = sir(y, beta, gamma, N)
        if beta_decay:
            beta = beta * (1 - beta_decay)
        s.append(S)
        i.append(I)
        r.append(R)

    s, i, r = np.array(s), np.array(i), np.array(r)
    return s, i, r


n_days = 183

beta_decay = 0.0

s, i, r = sim_sir(S, I, R, beta, gamma, n_days, beta_decay=beta_decay)


hosp = i * hospitalization_rate # TOTAL N of people needing hospitalization


days = np.array(range(0, n_days + 1))
data_list = [days, hosp]
data_dict = dict(zip(["day", "hosp"], data_list))

projection = pd.DataFrame.from_dict(data_dict)


# New cases
projection_admits = projection.iloc[:-1, :] - projection.shift(1)
projection_admits[projection_admits < 0] = 0

plot_projection_days = n_days - 10
projection_admits["day"] = range(projection_admits.shape[0])
```
